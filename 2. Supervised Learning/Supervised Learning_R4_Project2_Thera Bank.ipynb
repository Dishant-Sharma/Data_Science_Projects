<h1> Split the data into training and test set in the ratio of 70:30 respectively </h1>

```python
X=tb.drop('Personal Loan', axis=1)
y=tb['Personal Loan']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

<h1>Use different classification models (Logistic, K-NN and Naïve Bayes) to predict the likelihood of a liability customer buying personal loans (15 points)</h1>

```python
# As the datset is having a variety of columns , each having a different unit. so before applying these algorithms , standardising the data would be good. 
```

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
scaler.fit(X_train)

X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test)  
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
lg_model = LogisticRegression()
lg_model.fit(X_train, y_train)
```

```python
y_pred_lg = lg_model.predict(X_test)
```

```python
#KNN
```

```python
#Finding optimal value of k
```

```python
k=np.arange(1,20,2)
k
```

```python
accuracy_scores=[]
for i in k:
    model= KNeighborsClassifier(n_neighbors= i)
    model=model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    accuracy_scores.append(accuracy_score(y_test,y_pred))
```

```python
accuracy_scores #all the accuracy scores for k=1,3,...19
```

```python
# calculating mse
mse=[]
for i in accuracy_scores:
    mse.append(1-i)
```

```python
mse
```

```python
min(mse)
```

```python
mse.index(0.04466666666666663)
```

```python
#we can see that the lowest value of this error is when the value is equal to 5 so we implement KNN algorithm for k=5
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
pred_KNN = knn.predict(X_test)
```

```python
# Naive Bayes
```

```python
from sklearn.naive_bayes import GaussianNB

clf_GNB = GaussianNB()
clf_GNB = clf_GNB.fit(X_train, y_train)
```

```python
pred_GNB=clf_GNB.predict(X_test)
```

<h1>Print the confusion matrix for all the above models (5 points)</h1>

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix_lg = confusion_matrix(y_test, y_pred_lg)
print("The confusion matrix for logistic regression is :","\n",confusion_matrix_lg)
```

```python
confusion_matrix_knn = confusion_matrix(y_test, pred_KNN)
print("The confusion matrix for KNN is :","\n",confusion_matrix_knn)
```

```python
confusion_matrix_GNB = confusion_matrix(y_test, pred_GNB)
print("The confusion matrix for Naive Bayes is :","\n",confusion_matrix_GNB)
```

<h1>Give your reasoning on which is the best model in this case and why it performs better? (5 points)</h1>

```python
Logscore=lg_model.score(X_test, y_test)
print(Logscore)
KNNscore=knn.score(X_test, y_test)
print(KNNscore)
NBscore=clf_GNB.score(X_test,y_test)
print(NBscore)
```

```python
# As evident KNN gives the best accuracy of 95.53% . Also the confusion matrix of knn shows that more number of True posiitves and negatives and the sum of false positives and negatives is lower for the KNN model.
```

```python
#So,KNN is better to be implemented in this case.
```
